```python
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture
import plotly.express as px
```

```python
data = pd.read_csv("C:/Users/User/Documents/University/Internship/10-25-2021/Clustering Algorithms/train.csv")
data.head()
```

<h5>Preprocessing the dataset</h5>

```python
from sklearn.preprocessing import MinMaxScaler

# create a scaler object
scaler = MinMaxScaler()
# fit and transform the data
data_norm = pd.DataFrame(scaler.fit_transform(data.drop(['id','target'], 1)), columns=data.drop(['id','target'], 1).columns)

data_norm
```

```python
X = np.array(data_norm)
```

<h3>Initializing a Gaussian Mixture Model</h3>

```python
gmm = GaussianMixture(n_components=9, random_state=0)
gmm.fit(X)
```

```python
data["Cluster"] = gmm.predict(X)
```
