```python
# Dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.express as px
from sklearn.metrics import confusion_matrix
from sklearn.cluster import AffinityPropagation
```

```python
data = pd.read_csv("C:/Users/User/Documents/University/Internship/10-25-2021/Clustering Algorithms/train.csv")
data.head()
```

```python
data.isna().head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# create a scaler object
scaler = MinMaxScaler()

#taking a sample 2000 records from the data
data = data.sample(n=4000)

# fit and transform the data
data_norm = pd.DataFrame(scaler.fit_transform(data.drop(['id','target'], 1)), columns=data.drop(['id','target'], 1).columns)

data_norm
```

```python
data
```

```python
X = np.array(data_norm)
```

```python
y = np.array(data['target'])
```

```python
# Fit AFfinity Propagation with Scikit
afprop = AffinityPropagation(max_iter=250, random_state=0)
afprop.fit(X)
```

```python
# Predict the cluster for all the samples
data["Cluster"] = afprop.predict(X)
```
